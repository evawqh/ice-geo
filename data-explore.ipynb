```python
import pandas as pd
```

```python
df=pd.read_csv("clean_spend.csv")
df.head()
```

```python
state_year = df.groupby(by=["state","action_date"]).sum().reset_index()
state_year
```

```python
state_year.to_csv("state_year.csv", index=False)
```

```python
state_df = state_year.drop(columns=["action_date","total_dollars_obligated","potential_total_value_of_award"]).pivot_table(index="state", aggfunc=["count","sum","mean"]).reset_index()
state_df.columns =["state","num_years","sum","ave_year"]
state_df = state_df.sort_values(by="ave_year", ascending=False)
state_df
```

```python
state_df.to_csv("state_summary.csv", index=False)
```
